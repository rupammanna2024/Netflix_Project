```python


df = pd.read_csv('netflix_titles.csv')


df = df.dropna(subset=['type','release_year','rating','country','duration'])


type_counts = df['type'].value_counts()


plt.bar(type_counts.index, type_counts.values, color=['skyblue','orange'])


plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Distribution of Netflix Content (Movies vs TV Shows)')

plt.show()
```

```python
rating_count  = df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_count , labels=rating_count.index , autopct='%1.1f%%')
plt.tight_layout()
plt.title('percentage of Content rating')

plt.show()
```

```python
movie_df = df[df['type'] == 'Movie']
movie_df['duration_int'] = movie_df['duration']
plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'] ,bins=30, color='purple',edgecolor= 'black')
```

```python
release_count = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.scatter(release_count.index, release_count.values,marker ='o', color='red')
plt.grid(True)
plt.title('Release Year vs Number of Shows')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows')
plt.show()
```

```python
country_count = df['country'].value_counts().sort_index().head(10)
plt.barh(country_count.index,country_count.values,color = 'orange')
plt.xlabel('Number of shows')
plt.ylabel('Country')
plt.title('Top 10 counties by number of shows')
```
